```python
z_A, r, λ, ϕ = sym.symbols("z_A, r, λ, ϕ")
```

```python
pomik = sym.Eq(z_A, r*(1+(1/λ)-sym.cos(ϕ)-(1/λ)*sym.sqrt(1-λ**2*sym.sin(ϕ)*sym.sin(ϕ))))
pomik
```

```python
pomiki_gredi = np.zeros(360)
for i in range(0, 360):
    parametri = {r: 43, λ: lam, ϕ: radiani[i]}
    pomiki_gredi [i] = (pomik.rhs.subs(parametri))
```

## Grafični prikaz

```python
def prikazi_kote():
    return plt.xticks(np.arange(0, 721, 90))
```

```python
stopinje = np.arange(0, 720, 1)
radiani = stopinje * np.pi / 180
radij = 43
l = 134
lam = radij/l
```

```python
stopinje = np.arange(0, 720, 1)
radiani = stopinje * np.pi / 180
```

```python
def izracun_pomikov(ϕ_, r_ = radij, l_ = l):
    λ_ = r_ / l_
    return(r_*(1+(1/λ_)-np.cos(ϕ_)-(1/λ_)*(1-λ_**2*np.sin(ϕ_)*np.sin(ϕ_))**(1/2)))
pomiki = izracun_pomikov(radiani)
```

```python
def izracun_pomikov_graf(Radij = 43, Dolzina_gredi = 134):
    plt.plot(stopinje, izracun_pomikov(radiani, r_ = Radij, l_ = Dolzina_gredi))
    prikazi_kote()
    plt.grid()
    plt.xlabel("Zasuk ročične gredi [stopinje]")
    plt.ylabel("Pomik bata [mm]");
```

```python
interact(izracun_pomikov_graf, Radij = 
         np.arange(10, 100, 1), Dolzina_gredi = np.arange(50,200,1));
```

```python
mrtvi_volumen = np.pi*radij**2*l/9.5
mrtvi_volumen
```

```python
def racunaj_prostornino(r_ = 43, l_ = 134):
    return np.pi * r_ ** 2 *izracun_pomikov(radiani, r_, l_) + mrtvi_volumen
volumni = racunaj_prostornino()
```

```python
vrednosti = lines[22:-1]
tlaki = np.array([])
for i in vrednosti:
    tlaki = np.append(tlaki, float(i[2]))
vrednosti
```

```python
def p_V_diagram():
    plt.plot(racunaj_prostornino(), tlaki)
    plt.grid();
    plt.xlabel("Volumen [mm^3]")
    plt.ylabel("Tlak [bar]");
    plt.title("p(v) diagram")
p_V_diagram()    
```

```python
plt.plot(stopinje, tlaki)
plt.grid()
prikazi_kote()
plt.xlabel("Zasuk [Stopinje]")
plt.ylabel("Tlak [bar]")
plt.title("p(ϕ) diagram");
```

```python
def odvod_tlaka(p_, ϕ_):
    return np.gradient(p_, ϕ_, edge_order = 1)
```

```python
plt.plot(stopinje, odvod_tlaka(tlaki, radiani));
plt.grid()
prikazi_kote()
plt.xlabel("Zasuk [Stopinje]")
plt.ylabel("dp/dϕ [bar/rad]")
plt.title("Odvod tlaka v odvisnosti od zasuka");
```
