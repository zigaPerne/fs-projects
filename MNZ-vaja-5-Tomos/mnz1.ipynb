# Izračun moči motorja



Grafični prikaz

```python
stopinje = np.arange(0, 720, 1) #naredi tabelo od 0 do 719 stopinj
radiani = stopinje * np.pi / 180 #pretvorba v rad
radij = 43
l = 134
lam = radij/l #lambda
```

```python
def izracun_pomikov(ϕ_, r_ = radij, λ_ = lam):
    return(r_*(1+(1/λ_)-np.cos(ϕ_)-(1/λ_)*(1-λ_**2*np.sin(ϕ_)*np.sin(ϕ_))**(1/2))) #enacba za raunanje pomika
```

```python
plt.plot(stopinje, izracun_pomikov(radiani))
plt.xticks(np.arange(0, 730, 90))
plt.grid()
plt.xlabel("Zasuk ročične gredi [stopinje]")
plt.ylabel("Pomik bata [mm]");
```
